```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
data=data.drop("User ID",axis=1)
```

```python
data["Purchased"].value_counts()
```

```python
data.columns
```

```python
indep=data[['Age', 'EstimatedSalary','Gender_Male']]
data.shape
```

```python
dep=data[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
para_grid={'n_neighbors':[4,8,12,16],'weights':["uniform", "distance" , "callable"],'algorithm':["auto", "ball_tree", "kd_tree", "brute"]}
grid=GridSearchCV(KNeighborsClassifier(),para_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,y_train)
re=grid.cv_results_
y_pred=grid.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
clf=classification_report(y_test,y_pred)
```

```python
print("The R_score value for best parameter():",format(grid.best_params_),clf)
table=pd.DataFrame.from_dict(re)
table
```
